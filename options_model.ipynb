```python
def days_until_expiry(timestamp, expiry_date_str):
    timestamp = float(timestamp)
    proper_date = expiry_date_str.split("-")[1]
    # Convert the expiry date string to a datetime object
    expiry_date = datetime.datetime.strptime(proper_date, '%d%b%Y')

    # Convert the timestamp to a datetime object
    timestamp_dt = datetime.datetime.fromtimestamp(timestamp / 1e9)

    # Calculate the number of days between the timestamp and the expiry date
    days_until_expiry = (expiry_date - timestamp_dt).days

    # Return the number of days
    return days_until_expiry
```

## Add intrinsic value column
check if the option is a call or a put and calculate the intrinsic value

```python
def intrinsic(price,strike,put):
    price = float(price)
    strike = float(strike)
    if put:
        return abs(price-strike) if price<strike else 0
    return price-strike if price>strike else 0
```

```python
df
```

```python
df["intrinsic"] = df.apply(lambda row: intrinsic(row['price'],row['strike'],row["Put"]),axis=1)
```

```python
X=df.drop(['contract_id','filled_price','contract_label', 'date'],axis=1)
y=df['filled_price']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rfc = RandomForestRegressor(n_estimators=1000)
```

```python
rfc.fit(X_train,y_train)
```

```python
pred = rfc.predict(X_test)
```

```python
from sklearn import metrics

print('MAE:', metrics.mean_absolute_error(y_test, pred))
print('MSE:', metrics.mean_squared_error(y_test, pred))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, pred)))
```
